```python
df_data_amb=df_data_amb.dropna()
df_data_amb=df_data_amb.reset_index(drop=True)
df_data_amb
```

```python
#there are 254487 rows of data. Each comprise of 9 categories of 'Ambience'.
#To create one-hot encoding on this dataset in one batch took many hours.
#To overcome this, split the batch into smaller chunks and process them iteratively. 
#This took about 1 hour. Significantly less than processing it in one batch. 
```

```python
def chunker(k):
    for i in k:
        di=ds_list[i]
        d=ast.literal_eval(di)
        for n in d.keys():
            if d[n]==True:
                df_data_amb.loc[i,n]=1
            else :
                df_data_amb.loc[i,n]=0
    print("End :", k)
                
k= np.array_split(np.array(range(len(ds_list))),100)
#print (k[0])
for k_num in k:
    chunker(k_num)
    print ("Start :",k_num)
df_data_amb.head()
```

```python
#get categories list
categories_list={}
for i in range(len(df_data_amb)): 
    c=df_data_amb.categories[i]
    for n in list(c.split(',')):
        categories_list[n]=0
categories_list
```

```python
len(categories_list)
```

```python
#make each categories a column in dataset

zeroArr_Categories=np.zeros(len(df_data_amb)).astype(int)
for i in categories_list.keys():
    categories_list[i]=np.array(zeroArr_Categories)
categories_list
```

```python
#From the list of catergories, entries reference to F&B categories, type of business concept, lifestyle concept etc.
#a total of 122 categories were identified. 
```

```python
categorie_list_df= pd.DataFrame(categories_list)
categorie_list_df.head()
```

```python
df_data_merged_categories=df_data_amb.join(categorie_list_df, how='outer')
df_data_merged_categories.head()
```

```python
df_data_merged_categories.shape
```

```python
import time

#implement chunking to reduce the processing time. 
def chunker_categories(k,k_index):
    start = time.time()
    for i in k:
        CatStr=df_data_merged_categories.loc[i,'categories']
        for n in list(CatStr.split(',')):
            df_data_merged_categories.loc[i,n]=1
    end = time.time()
    print("Completed","Time used:", end-start,' : ', k_index, 'of 100 : ', k)
                
k= np.array_split(np.array(range(len(df_data_merged_categories))),100)
for index,k_num in enumerate(k):
    print ("Start :",index+1,' of 100', ' : ',k_num)
    chunker_categories(k_num, index+1)

```
