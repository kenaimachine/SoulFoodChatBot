```python
import pandas as pd
import numpy as np

# read the entire file into a python array
with open('yelp_dataset/review.json', 'r') as f:
    data = f.readlines()

# remove the trailing "\n" from each line
data = map(lambda x: x.rstrip(), data)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_json_str = "[" + ','.join(data) + "]"

# now, load it into pandas
data_df = pd.read_json(data_json_str)
data_df.head(10)
```

```python
# read the entire file into a python array
with open('yelp_dataset/business.json', 'r') as f:
    data_business = f.readlines()

# remove the trailing "\n" from each line
data_business = map(lambda x: x.rstrip(), data_business)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_business_json_str = "[" + ','.join(data_business) + "]"

# now, load it into pandas
data_df_business = pd.read_json(data_business_json_str)
data_df_business.head(10)
```

```python
import ast
def getAttributeList(attributeDictString):
    d=ast.literal_eval(attributeDictString)
    return d.keys()
```

```python
s=df_data_amb.attributes
from pandas.io.json import json_normalize
AttriList={}

sdf=pd.DataFrame(json_normalize(s))
sdf

    
```

```python
for i in range(len(s)):
    for j in getAttributeList(sdf.Ambience[0]):
        AttriList[j]=0
AttriList
```

```python
zeroArr_Ambience=np.zeros(len(df_data_amb)).astype(int)
for i in AttriList.keys():
    AttriList[i]=np.array(zeroArr_Ambience)
AttriList
```
